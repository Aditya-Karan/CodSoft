```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix
```

```python
# seprating target column from feature columns
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#splitting train and testing data from our data by using sklearn's train_test_split
#split propotion is set to default at 75% and 25% respectively
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

#prints dimension of training and testing data
print("Shape of Training Data : " + str(X_train.shape))
print("Shape of Testing Data : " + str(X_test.shape))
```

```python
#training our DecisionTreeClassifier with training data and adjusting parameters to avoid overfitting
model = DecisionTreeClassifier(max_depth = 4).fit(X_train, y_train)

#print accuracy score on training and testing data
print("Model Accuracy on Training Data : " + str(model.score(X_train, y_train)))
print("Model Accuracy on Testing Data : " + str(model.score(X_test, y_test)))

#plots tree to visualize how did we split data
print("Plotting Split of Our Decision Tree")
plot_tree(model)

predicted_y = model.predict(X_test)

confusion_matrix(predicted_y, y_test)
```
